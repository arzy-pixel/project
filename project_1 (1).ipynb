```python
import pandas as pd

df = pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
print(df.head())
```

```python
print('Первые 5 строк:')
print(df.head())

print('Последние 5 строк:')
print(df.tail())
```

```python
df.info()
```

```python

df['Sleep Duration'] = df['Sleep Duration'].astype(int)

df.info()
```

```python
df.isna()
```

```python
df.isnull().sum()
```

```python

df.dropna(subset=['Sleep Disorder'], inplace=True)
print(df)
```

```python
duplicate_rows = df.duplicated().sum()
print(f"количество дублирующихся строк: {duplicate_rows}")
```

```python
print(df.nunique())
```

```python
print(df['Sleep Disorder'].value_counts())
```

```python

correlation= df.corr(numeric_only=True)
print(correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Age'], bins=20)
plt.title('Распределение возраста участников')
plt.show()
```

```python

age_sleep = df.groupby('Age')['Sleep Duration'].mean()
sns.lineplot(x=age_sleep.index, y=age_sleep.values)
plt.title('Изменение длительности сна с возрастом')
plt.ylabel('Средний сон (часы)')
plt.show()
```

```python
sns.countplot(x='Occupation', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Gender', y='Quality of Sleep', data=df)
plt.title('Сравнение качества сна у мужчин и женщин')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Матрица корреляции всех параметров')
plt.show()
```

```python
# Группируем по полу и считаем среднее для числовых колонок
avg_by_gender = df.groupby('Gender').mean(numeric_only=True)
print(avg_by_gender)
```

```python

prof_stats = df.groupby('Occupation').agg({
    'Quality of Sleep': 'mean',
    'Age': 'mean',
    'Sleep Duration': 'max'
})
print(prof_stats)
```

```python

pivot = df.pivot_table(index='Occupation', columns='Gender', values='Sleep Duration', aggfunc='mean')
print(pivot)
```
